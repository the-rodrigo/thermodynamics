```python
import math
import numpy
import matplotlib.pyplot as plt

def get_var_A(vf:int or float) -> float:
    var_A = -R*T*(n_b*math.log((v_total-vf)/(v_total-v_ref)) + n_a*math.log(vf/v_ref))
    return var_A

n_a: int or float = 1 #mol
n_b: int or float = 3 #mol
v_ref: float = 0.02 #m^3
R: float = 8.314 #J/(mol*K)
T: int or float = 298 #K
v_total: int or float = 0.1 #m^3
Da:list[int or float] = []

v_final = numpy.linspace(0.02,0.03,100)

for vf in v_final:
    var_A = get_var_A(vf)
    Da.append(var_A)

va = 0.025 #m^3
Da_min = get_var_A(va)

fig, ax = plt.subplots()
ax.plot(v_final,Da,'r-',va,Da_min,'bo')
plt.title('Gráfico de Variação de Helmholtz x Volume')
plt.xlabel("Vf (m^3)")
plt.ylabel("Delta_A (J)")
plt.show()
```
